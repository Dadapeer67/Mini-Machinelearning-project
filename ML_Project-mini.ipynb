```python
X=iris.drop(['species'],axis=1)
```

```python
Y=iris['species']
```

```python
Y
```

```python
le=LabelEncoder()

```

```python
le=LabelEncoder()
Y=le.fit_transform(Y)
Y
```

ADD  RANDOM TEST TO TEST TRAINSPLIT

```python


x_train , x_test , y_train ,y_test=train_test_split(X,Y,test_size=0.4)
```

Decesion trees

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(solver='newton-cg')
```

```python
lr.fit(x_train,y_train)
```

```python
y_predi1=lr.predict(x_test)
```

```python
confusion_matrix(y_test,y_predi1)
```

```python
sns.heatmap(data=confusion_matrix(y_test,y_predi1),annot=True,cmap='Blues')
```

```python
accuracy_score(y_test,y_predi1)
```

DECESION TREE

```python
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
y_pred2=dtree.predict(x_test)
```

confusion matrix and heat map

```python
confusion_matrix(y_test,y_pred2)
```

```python
sns.heatmap(data=confusion_matrix(y_test,y_pred2),annot=True,cmap='Blues')
```

```python
accuracy_score(y_test,y_pred2)
```

```python
iris.columns[:-1]
```

```python
plt.figure(figsize=(20,20))
dec_tree=plot_tree(decision_tree=dtree,feature_names=iris.columns[:-1],
 class_names=["setosa","vercicoloar","verginica"],filled=True)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=1000)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred3=clf.predict(x_test)
```

get confusion matrix and heatmap

```python
confusion_matrix(y_test,y_pred3)
```

```python
sns.heatmap(data=confusion_matrix(y_test,y_pred3),annot=True,cmap='Blues')
plt.ylabel('Actual label')
plt.xlabel('predicted label')
```

```python
accuracy_score(y_test,y_pred3)
```
